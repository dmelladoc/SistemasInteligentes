# K vecinos

Este es un ejemplo de implementación de K-vecinos

Vamos a cargar las librerías `numpy y matplotlib`


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Primero vamos a crear unos datos
np.random.seed(10)
X1 = np.random.normal(2, 0.5, (100,2))
X2 = np.random.normal(-1, 0.8, (100,2))
Y = np.array([0]*100 + [1]*100)

X = np.append(X1,X2,axis=0)
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
dato = np.array([0,0.5])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(dato[0],dato[1],marker="x")
```

```python
k = 10 #10 vecinos mas cercanos
#medir la distancia euclideana
dist = np.sqrt(np.sum((X - dato)**2,axis=1))
print(dist)
```

```python
ordenado = np.argsort(dist)
print(ordenado)
print(Y[ordenado[:k]])
```

# Perceptron


```python
W = np.random.rand(1,2)
print(W)
```

```python
#parametros
lr = 0.1

#prediccion
yhat = np.dot(W,X.T)
print(yhat)
```

```python
yhat = np.where(yhat>0,1,0)
print(yhat)
```

```python
error = Y - yhat
print(error)
```

```python
correccion = lr * error.sum()
print(correccion)
```

```python
W = W + X.sum(0)*correccion
print(W)
```

```python
exactitud = np.sum(yhat == Y)/200
print(exactitud)
```

# Regresion Logistica

```python
W = np.random.rand(1,2)
print(W)
```

```python
def sigmoide(x):
    return 1/(1+np.exp(-x))
```

```python
z = np.dot(W,X.T)
pred = sigmoide(z)
print(pred)
```

```python
error = Y - pred
gradiente = np.dot(X.T,error.T)
print(gradiente)
W += 0.001*gradiente.T
print(W)
```

```python
yhat = sigmoide(np.dot(W,X.T))
pred = np.where(yhat >= 0.5, 1, 0)
exactitud = np.sum(pred == Y)/200
print(exactitud)
```
